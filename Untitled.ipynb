```python
#!/usr/bin/env python3

import csv
import numpy as np
import functools 
import sys


input_file = sys.argv[1] #"/Users/romanvyjidacek/phd/datasets/csv/martin/firewall1.csv"

with open(input_file, 'r') as csvfile:
    np_matrix = np.array(list(csv.reader(csvfile, delimiter=','))).T

matrix = np_matrix.tolist()

enumerated = list(enumerate(matrix))

enumerated.sort(reverse=True, key=lambda t: functools.reduce(lambda result, element: int(result) + int(element), t[1]))

#sorted_context = [[0 for _ in range(len(enumerated))] for _ in range(len(enumerated[0][1]))]

permutation = list(map(lambda e: e[0], enumerated))
sorted_context = np.array(list(map(lambda e: e[1], enumerated))).T.tolist()

print(permutation)
print()

for row in sorted_context:
    print(','.join(map(str, row)))
    
```
